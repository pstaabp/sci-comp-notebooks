## Chapter 26: Differential Equations

```julia
using DifferentialEquations, Plots
```

```julia
f(u,p,t) = u
u0 = 1.0
tspan = (0.0,1.0)
prob = ODEProblem(f,u0,tspan)
```

```julia
sol = solve(prob)
```

```julia
[(t,sol(t)) for t=0:0.1:1]
```

```julia
plot(sol,lw=3)
plot!(0:0.01:1,map(exp,0:0.01:1),lw=3,ls=:dash)
```

```julia
plot(0:0.01:1,map(t->sol(t)-exp(t),0:0.01:1))
```

```julia
function F(du,u,p,t)
  du[1] = u[2]
  du[2] = - u[1]
end
```

```julia
prob2 = ODEProblem(F,[0;1],(0,6pi))
```

```julia
sol2 = solve(prob2)
```

```julia
plot(sol2)
```

```julia
plot(sol2,vars=(1,2),aspect_ratio=:equal)
```

```julia
function lorenz!(du,u,p,t)
 du[1] = 10.0*(u[2]-u[1])
 du[2] = u[1]*(28.0-u[3]) - u[2]
 du[3] = u[1]*u[2] - (8/3)*u[3]
end
```

```julia
u0 = [1.0;0.0;0.0]
tspan = (0.0,100.0)
prob3 = ODEProblem(lorenz!,u0,tspan)
sol3 = solve(prob3)
```

```julia
plot(sol3)
```

```julia
plot(sol3,vars=(1,2,3))
```
