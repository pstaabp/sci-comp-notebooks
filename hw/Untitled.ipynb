```julia
function fact(n::Integer)
  prod = 1
  for k=1:n
    prod *= k
  end
  @show prod
end
```

```julia
fct = fact(8)
```

```julia
function triangle(x::Real)
  if 0 < mod(abs(x),2) < 1
    return mod(abs(x),2)
  else 
    return 0.5
  end
end
```

```julia
plot(triangle,-4.5,4.5)
```
